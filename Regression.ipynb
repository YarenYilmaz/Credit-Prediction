```python
# TRAIN & TEST DATA
from sklearn.model_selection import train_test_split
X = df_new.drop(["creditAmount"], axis = 1)
y = df_new["creditAmount"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
# NORMALIZE DATA
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X_train.values)
X_train = pd.DataFrame(x_scaled, columns = X_train.columns)

x_scaled = min_max_scaler.fit_transform(X_test.values)
X_test = pd.DataFrame(x_scaled, columns = X_test.columns)
```

```python
# KNN K = 3

model_knn_3 = KNeighborsRegressor(n_neighbors=3)
model_knn_3.fit(X_train, y_train)
```

```python
# KNN K = 4

model_knn_2 = KNeighborsRegressor(n_neighbors=4)
model_knn_2.fit(X_train, y_train)
```

```python
# ERROR CALCULATION K = 3

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from math import sqrt

predicted_x_train = model_knn_3.predict(X_train)
mse = mean_squared_error(y_train, predicted_x_train)
mae = mean_absolute_error(y_train, predicted_x_train)
print("mse = ",mse," & mae = ",mae," & rmse = ", sqrt(mse))

predicted_x_test = model_knn_3.predict(X_test)
test_mse = mean_squared_error(y_test, predicted_x_test)
test_mae = mean_absolute_error(y_test, predicted_x_test)
print("mse = ",test_mse," & mae = ",test_mae," & rmse = ", sqrt(test_mse))
```

```python
# ERROR CALCULATION K = 4

predicted_x_train = model_knn_2.predict(X_train)
mse = mean_squared_error(y_train, predicted_x_train)
mae = mean_absolute_error(y_train, predicted_x_train)
print("mse = ",mse," & mae = ",mae," & rmse = ", sqrt(mse))

predicted_x_test = model_knn_2.predict(X_test)
test_mse = mean_squared_error(y_test, predicted_x_test)
test_mae = mean_absolute_error(y_test, predicted_x_test)
print("mse = ",test_mse," & mae = ",test_mae," & rmse = ", sqrt(test_mse))
```
